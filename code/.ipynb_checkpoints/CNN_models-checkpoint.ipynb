## Word Embedding

```
# load the whole embedding into memory
embeddings_index = dict()
f = open(ROOT_PATH + 'resource/cc.vi.300.vec')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('Loaded %s word vectors.' % len(embeddings_index))
```

```
def make_padded_docs(texts, max_length, tokenizer):
    tokenizer.fit_on_texts(texts)
    vocab_size = len(tokenizer.word_index) + 1
    # integer encode the documents
    encoded_docs = tokenizer.texts_to_sequences(texts)
    # print(encoded_docs)
    # pad documents to a max length of 4 words
    padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
    print(f'padded_docs shape: {padded_docs.shape}')
    padded_docs

    return padded_docs, vocab_size
```

```
def make_weight_matrix(embeddings_index, vocab_size, dim, tokenizer):
    # create a weight matrix for words in training docs
    embedding_matrix = np.zeros((vocab_size, dim))
    for word, i in tokenizer.word_index.items():
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
    print(f'embedding_matrix shape: {embedding_matrix.shape}')

    return embedding_matrix
```

```
max_length = 128
dim = 300
# prepare tokenizer
tokenizer = Tokenizer()
padded_docs, vocab_size = make_padded_docs(train.content.tolist(),
                                                 max_length,
                                                 tokenizer)
embedding_matrix = make_weight_matrix(embeddings_index,
                                            vocab_size,
                                            dim,
                                            tokenizer)
```

```
labels = np.array(train.ground)
Counter(labels)
```

```
X_train, X_val, y_train, y_val = train_test_split(padded_docs, labels, test_size=0.2)
```

## CNN_LSTM model

```
from keras.models import Sequential
from keras.layers import Dense, Input, Bidirectional, LSTM, RepeatVector, InputLayer
from keras.layers import Flatten
from keras.layers import Embedding
from keras.layers import Dropout
from keras.layers import Conv1D
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras import regularizers
```

```
# define model
model = Sequential()
embedding_layer = Embedding(vocab_size, dim, weights=[embedding_matrix], input_length=max_length, trainable=False)
model.add(embedding_layer)
model.add(Dropout(0.2))
model.add(Conv1D(2048, 7, activation='relu'))
model.add(GlobalMaxPooling1D())
# model.add(Dropout(0.5))
# model.add(Dense(64, kernel_regularizer=regularizers.l2(0.001), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'],)
# plot the model
plot_model(model, to_file=ROOT_PATH + 'resource/CNNLSTM_1231.png', show_shapes=True, show_layer_names=True)
```

## Train

```
# fit the model
history = model.fit(X_train, y_train,
                    validation_data=(X_val, y_val),
                    batch_size=256,
                    epochs=10, verbose=1)
```
